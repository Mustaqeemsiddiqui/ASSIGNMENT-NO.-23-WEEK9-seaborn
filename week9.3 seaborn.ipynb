**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

**ANSWER**

Seaborn is a powerful Python visualization library based on matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics.

1. **Scatter Plot (sns.scatterplot)**
   - **Use**: A scatter plot displays values for typically two variables for a set of data. It is used to observe relationships between variables, showing how one variable is affected by another. It can also help identify outliers and trends within the data.
   - **Example**: Analyzing the relationship between height and weight in a dataset of individuals.

2. **Bar Plot (sns.barplot)**
   - **Use**: A bar plot represents data with rectangular bars, where the lengths of the bars are proportional to the values they represent. It is useful for comparing the values of different groups or categories.
   - **Example**: Comparing the average scores of different groups in a survey.

3. **Histogram (sns.histplot)**
   - **Use**: A histogram is used to represent the distribution of a dataset. It divides the data into bins and counts the number of observations in each bin. This is useful for understanding the underlying frequency distribution of data.
   - **Example**: Visualizing the distribution of ages in a population sample.

4. **Box Plot (sns.boxplot)**
   - **Use**: A box plot (or whisker plot) displays the distribution of data based on a five-number summary: minimum, first quartile, median, third quartile, and maximum. It is useful for identifying outliers and comparing distributions between different groups.
   - **Example**: Comparing the salaries across different job titles.

5. **Heatmap (sns.heatmap)**
   - **Use**: A heatmap is used to show the magnitude of a phenomenon as color in two dimensions. It is often used to display the result of a matrix, such as a correlation matrix, where relationships between multiple variables can be easily visualized.
   - **Example**: Visualizing the correlation between different features in a dataset.

```python

import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))  
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')  # annot=True to show the numbers, fmt='d' format

# Customize labels and title
plt.title('Number of Passengers (1000s) per Month')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```
